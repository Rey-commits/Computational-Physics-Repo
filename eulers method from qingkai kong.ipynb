```python
# Euler's Method

import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-poster")
%matplotlib inline
g = 9.81
c = 12.5
m = 68.1
# Define parameters
f = lambda t, v: g-(c/m)*v # ODE
h = 0.2 # Step size
t = np.arange(0, 2 + h, h) # Numerical grid
v0 = 0 # Initial Condition
# Explicit Euler Method
v = np.zeros(len(t))
v[0] = v0
for i in range(0, len(t) - 1):
    v[i + 1] = v[i] + h*f(t[i], v[i])
    
print("t =",t)
print("v(t) =",v)
plt.figure(figsize = (12, 8))
plt.plot(t, v, "o-", label="Approximate/Euler")
plt.plot(t, g*(m/c)*(1-np.exp(-(c/m)*t)), "r", label="Exact")
plt.title("Approximate and Exact Solution for Simple ODE")
plt.xlabel("t")
plt.ylabel("v(t)")
plt.grid()
plt.legend(loc="lower right")
plt.show()
```
