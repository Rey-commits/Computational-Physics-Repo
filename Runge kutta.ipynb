```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp
plt.style.use("seaborn-poster")
%matplotlib inline
g = 9.81
c = 12.5
m = 68.1
h = 0.2
F = lambda t, v: g-(c/m)*v
t_eval = np.arange(0,2+h,h)
sol = solve_ivp(F, [0,2+h,h],[0] , t_eval=t_eval)
print("Look for the values of y:")
print("Note: v(t) and y values are the same")
print(sol)

plt.figure(figsize = (12,8))

# t and y arrays are the corresponding values of t and sol retrieve below.
t = np.array([0. , 0.2, 0.4, 0.6, 0.8, 1. , 1.2, 1.4, 1.6, 1.8, 2. ])
y = np.array([ 0.        ,  1.92642411,  3.78341235,  5.57346314,  7.2989889 ,
         8.96231784, 10.56569446, 12.11128035, 13.60115344, 15.03732086,
        16.42172047])
plt.plot(t,y, "-o", label="Approximate/Runge-Kutta")
plt.plot(t, g*(m/c)*(1-np.exp(-(c/m)*t)), "r", label="Exact")
plt.title("Approximate and Exact Solution for Simple ODE")
plt.xlabel("t")
plt.ylabel("v(t)")
plt.grid()
plt.legend(loc="lower right")
plt.show()
```
