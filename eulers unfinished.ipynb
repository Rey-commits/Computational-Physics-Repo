```python
import matplotlib.pyplot as plt
m = 1
c = 1
g = 9.8
y1_0 = 0
y2_0 = 20
g2_i = g - (c/m)*y2_0
y_t_result = []
v_t_result = []
t_stepSize = [0,0.2,0.4,0.6,0.8,0.9,1,1.2,1.4,1.6,1.8,2]
for n in t_stepSize:
    y_t = y1_0 + n*y2_0
    v_t = y2_0 + n*g2_i
    y_t_result.append(y_t)
    v_t_result.append(v_t)
    
print("y(t) =",y_t_result)
print("v(t) =", v_t_result)
plt.plot(t_stepSize, y_t_result)
```
